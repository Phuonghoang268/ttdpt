```python
import numpy as np
import scipy.fftpack as fft
from math import pi
import matplotlib.pyplot as plt
```

```python
f=500
f_base=50
f_sampling=8000

t=np.arange(0,1./f_base, 1./f_sampling)
s=np.sin(2*pi*f*t)
plt.plot(t,s)
plt.show()
```

```python
f=500
f_base=50
f_sampling=8000

t=np.arange(0,1./f_base, 1./f_sampling)
s=np.sin(2*pi*f*t+pi/2)

dft=fft.rfft(s)
p=plt.subplot(2,1,1)
p.plot(t,s)

p=plt.subplot(2,1,2)
p.plot(np.arange(t.size)*f_base,dft)
plt.show()
```

```python
f=500
f_base=50
f_sampling=8000

t=np.arange(0,1./f_base, 1./f_sampling)
s=np.sin(2*pi*f*t)

dft=fft.rfft(s)
dft1=fft.fft(s)
fig=plt.figure(figsize=(7,7))
p=plt.subplot(3,1,1)
p.plot(t,s)

p=plt.subplot(3,1,2)
p.plot(np.arange(t.size)*f_base,dft)

p=plt.subplot(3,1,3)
p.plot(np.arange(t.size)*f_base,np.imag(dft1))
plt.show()
```

```python
f=100
f_base=50
f_sampling=8000
t=np.arange(0,1./f_base,1./f_sampling)
s=np.sin(2*pi*f*t)

dft=fft.dct(s)
dft1=fft.fft(s)
fig=plt.figure(figsize=(7,7))

p=plt.subplot(3,1,1)
p.plot(t,s)

##DCT
p=plt.subplot(3,1,2)
p=plt.plot(np.arange(t.size)*f_base,dft)

#FFT
p=plt.subplot(3,1,3)
p.plot(np.arange(t.size)*f_base,np.imag(dft1))
plt.show()
```

```python
f=500
f_base=50
f_sampling=8000
t=np.arange(0,1./f_base,1./f_sampling)
s=np.sin(2*pi*f*t + pi/2)

dft=fft.rfft(s)
dft1=fft.fft(s)
fig=plt.figure(figsize=(10,10))

p=plt.subplot(3,1,1)
p.plot(t,s)

##DCT
p=plt.subplot(3,1,2)
p=plt.plot(np.arange(t.size)*f_base,dft)

#FFT phn o
p=plt.subplot(3,1,3)
p.plot(np.arange(t.size)*f_base,np.imag(dft1)) #### phn o

plt.show()
```

```python
f=500
f_base=50
f_sampling=8000
t=np.arange(0,1./f_base,1./f_sampling)
s=np.sin(2*pi*f*t + pi/2)

dft=fft.rfft(s)
dft1=fft.fft(s)
fig=plt.figure(figsize=(10,10))

p=plt.subplot(3,1,1)
p.plot(t,s)

##DCT
p=plt.subplot(3,1,2)
p=plt.plot(np.arange(t.size)*f_base,dft)

#FFT phn thc
p=plt.subplot(3,1,3)
p.plot(np.arange(t.size)*f_base,np.real(dft1)) #### phn thc
plt.show()
```
