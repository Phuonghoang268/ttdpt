# Entropy of image


```python
import numpy as np

def entropy(im):
    
    #compute normalized histogram -> p(g)
    
    # dùng để xác định số lượng các mức xám. ví dụ ở mức 0 là 379423, ở mức 255 là 70.
    p= np.array([(im==v).sum() for v in range(256)])
    
    # tìm tỉ lệ mức xám
    p=p/p.sum()
    
    # e=-sum (p(g)* log2(p(g))
    e=-(p*np.log2(p)).sum()
    
    return e
```

```python
%matplotlib notebook
from matplotlib import pyplot as plt
from skimage.io import imread, imshow, imsave
```

```python
im=imread("lena.jpg")
plt.figure()
imshow(im)
plt.axis("off")
plt.title(entropy(im))
plt.show()
```
