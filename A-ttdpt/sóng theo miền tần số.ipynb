```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f=2 # => T=1/2
period=[0.,2.]      # Trong 2s
N=500            
step=(period[1]-period[0])/N
t=np.arange(period[0], period[1], step)      # tạo mảng cách đều vs step, bắt đầu từ pe0-> pe1
s=np.sin(2 * np.pi * f* t)     #s=2pi*f*t
fig=plt.plot(t,s)
plt.show()
```

```python
f=1 # => T=1/2
period=[0.,2.]      # Trong 2s
N=500            
step=(period[1]-period[0])/N
t=np.arange(period[0], period[1], step)      # tạo mảng cách đều vs step, bắt đầu từ pe0-> pe1
N_COMP=7
s=np.zeros((N,))
for k in range(0,N_COMP):
    s=s+np.sin(2 * np.pi * f * (2*k+1) * t)/(2*k+1)
plt.plot(t,s)
plt.show()
```
