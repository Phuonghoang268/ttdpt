## Introduction

Ngày nay, dữ liệu và nhu cầu truyền dữ liệu trở thành một phần quan trọng trong đời sống hàng ngày. Hình ảnh là một trong những đại diện phổ biến nhất của dữ liệu. Hình ảnh kỹ thuật số được sử dụng cho mục đích chuyên nghiệp và cá nhân, từ tài liệu chính thức đến phương tiện truyền thông xã hội. Do đó, bất kỳ tổ chức hoặc cá nhân cần lưu trữ và chia sẻ một số lượng lớn hình ảnh. Các thiết bị chụp ảnh ngày càng đa dạng hóa và có có khả năng chụp ảnh chất lượng càng cao.

Một trong những vấn đề phổ biến nhất liên quan đến việc sử dụng hình ảnh là kích thước hình ảnh. Các công nghệ thu hình và sự gia tăng thông tin dạng số đồng nghĩa với việc hình ảnh hiện có độ phân giải và chất lượng rất cao, dẫn đến sự gia tăng kích thước. Vì vậy nén hình ảnh đã trở thành một trong những phần quan trọng nhất của xử lý hình ảnh ngày nay.

Mục tiêu của nén ảnh là khiến ảnh đạt được kích thước nhỏ nhất có thể mà không ảnh hưởng đến chất lượng của hình ảnh. Ở đây, em sẽ nói về  thuật toán Run-length coding sử dụng để nén hình ảnh nhị phân.


## Background and related work

#### Nhu cầu nén ảnh

Trong một số ngành nhất định, việc lưu trữ số lượng lớn hình ảnh là bắt buộc. Một ví dụ điển hình là ngành y tế, nơi diễn ra quá trình quét hoặc lưu trữ liên tục các hình ảnh và tài liệu y tế. Việc lưu trữ khối lượng lớn hình ảnh đặt ra một thách thức lơn, đòi hỏi nhiều băng thông hơn. Kĩ thuật nén ra đời giúp lưu trữ không gian tốt hơn và truyền dữ liệu tốt hơn. 

Ví dụ, nếu một hình ảnh ở định dạng tệp không nén được truyền đi, chẳng hạn như định dạng BMP phổ biến, thì kích thước tệp có thể rất lớn và việc truyền hình ảnh qua internet sẽ mất quá nhiều thời gian để hoàn thành. Cụ thể, một hình ảnh có độ phân giải pixel là 1024 x 1024 pixel và độ phân giải màu 24 bit sẽ chiếm 1024 * 1024 * 24/8 = 3.145.728 (3MB) dung lượng lưu trữ và 7 phút để truyền, sử dụng tốc độ cao của 64Kbit/s, trong một dòng ISDN. Mặt khác, nếu hình ảnh được nén ở tỷ lệ nén 10:1, thì yêu cầu về dung lượng lưu trữ chỉ giảm xuống còn 300KB và thời gian truyền giảm xuống dưới 6 giây [5].

Nén là một thành phần quan trọng của các giải pháp có sẵn để tạo kích thước tệp có kích thước có thể quản lý và truyền được. Tăng băng thông là một phương pháp khác, nhưng chi phí đôi khi làm cho giải pháp này kém hấp dẫn hơn. Tính di động và hiệu suất của nền tảng rất quan trọng trong việc lựa chọn kỹ thuật nén/giải nén sẽ được sử dụng. Cách dễ nhất để giảm kích thước của tệp hình ảnh là giảm kích thước của chính hình ảnh đó. Bằng cách thu nhỏ kích thước của hình ảnh, cần lưu trữ ít pixel hơn và do đó, tệp sẽ mất ít thời gian hơn để tải [7]



#### Nén ảnh

Nén ảnh là một phần cơ bản của xử lý ảnh. Hình ảnh chất lượng cao thường có kích thước lớn và do đó việc truyền những hình ảnh này chậm hơn và cần nhiều băng thông hơn. Một tệp lớn cũng yêu cầu nhiều không gian lưu trữ hơn trên máy chủ và do đó giới hạn số lượng hình ảnh có thể được lưu trữ. Quá trình nén đạt được bằng cách loại bỏ một hoặc nhiều trong ba dữ liệu dư thừa cơ bản[ 1]
   * Dư thừa mã: Sự dư thừa bằng cách sử dụng các từ mã có độ dài khác nhau để biểu diễn một cách không mất dữ liệu các symbol có xác suất khác nhau được gọi là sự dư thừa mã. Sử dụng số bit nhỏ để biểu thị các ký hiệu thường xuyên hơn và sử dụng số lượng lớn bit để biểu thị các ký hiệu ít phổ biến hơn để có thể giảm độ dài của tổng thể các symbol.
   
   * Dư thừa sự liên kết các pixel: là sự dư thừa tương ứng với sự phụ thuộc thống kê giữa các pixel, đặc biệt là giữa các pixel liền kề. Hầu hết các mảng cường độ 2-D có tương quan không gian trong hình ảnh và thông tin không cần thiết phải lặp lại trong biểu diễn. Thông tin lặp đi lặp lại này là dư thừa và được khai thác cho mục đích nén.
   
   * Dư thừa tâm lý thị giác: là sự dư thừa tương ứng với độ nhạy nhận thức khác nhau đối với tất cả các tín hiệu hình ảnh do mắt người tạo ra. Do đó, việc loại bỏ một số thông tin ít tương đối quan trọng hơn trong quá trình xử lý hình ảnh có thể được chấp nhận

#### Phân loại nén ảnh
Nén hình ảnh có thể chia làm 2 loại: nén không mất dữ liệu ( Lossless compression) và nén mất dữ liệu (Lossy compression) [6].

* Nén mất dữ liệu : là làm giảm kích thước bằng cách loại bỏ một số phần của hình ảnh mà những phần đó không cần thiết lắm. Kỹ thuật này làm hỉnh ảnh nhỏ hơn, và sự thay đổi đáng kể so với ảnh gốc. Nó được sử dụng khi việc tái tạo lại chính xác ảnh gốc không quan trọng lắm, nhưng với kì vọng tỉ lệ nén cao.

* Nén không mất dữ liệu: là thay đổi kích thước hình ảnh thành một hình ảnh như bản sao chính xác của hình ảnh với kích thước nhỏ hơn. Kỹ thuật này không làm ảnh hưởng đến chất lượng, mất mát thông tin của hình ảnh.Trong các kỹ thuật nén không mất dữ liệu, hình ảnh gốc có thể được phục hồi hoàn hảo từ hình ảnh được nén (mã hóa). Chúng còn được gọi là không nhiễu vì chúng không thêm nhiễu vào tín hiệu (hình ảnh). Nó còn được gọi là mã hóa entropy vì nó sử dụng các kỹ thuật thống kê/phân tách để loại bỏ/giảm thiểu sự dư thừa. Nén không mất dữ liệu chỉ được sử dụng cho một số ứng dụng có yêu cầu nghiêm ngặt như hình ảnh y tế. Các kĩ thuật nén không mất dữ liệu phỏ biến: Run length encoding, Huffman encoding , LZW coding , Area coding

#### Run-length encoding 
Run-Length Encoding (RLE) là thuật toán nén không mất dữ liệu cơ bản và đơn giản. Ý tưởng đằng sau cách tiếp cận RLE để nén dữ liệu là: nếu một mục dữ liệu d xuất hiện n lần liên tiếp trong luồng đầu vào, hãy thay thế n lần xuất hiện bằng một cặp duy nhất, n lần xuất hiện liên tiếp của một mục dữ liệu được gọi là độ dài chạy n (run length of n) [2].



## Details of Your topic of interest




Run-length encoding (RLE) dựa trên việc tìm các lần chạy dữ liệu, tức là chuỗi các phần tử dữ liệu tương tự trong luồng dữ liệu đầu vào và thay thế các chuỗi lặp lại bằng một giá trị phần tử dữ liệu duy nhất bằng số đếm của nó.[3]


![loss](lossless_RLE.jpg " nén các pixel liên tiếp bằng cách chỉ thay thế mỗi lần chạy bằng một pixel từ nó và một bộ đếm hiển thị số lượng mục mà nó chứa.")


Ví dụ ta chạy một chuỗi với với B đại diện cho một pixel màu đen và W đại diện cho màu trắng

WWWWWWWWWWWWBWWWWWWWWWWWWBBBWWWWWWWWWWWWWWWWWWWWW
WWWBWWWWWWWWWWWWWW

Nếu ta áp dụng thuật toán run-length encoding (RLE), chúng ta sẽ nhận được kết quả sau: 12W1B12W3B24W1B14W. Đây là một kỹ thuật chung để mã hóa các ký hiệu dựa trên tần suất xuất hiện thống kê (xác suất) của chúng.

RLE vừa dễ thực hiện vừa nhanh chóng, hơn thế nữa, nó yêu cầu rất ít kích thước bộ nhớ và dễ giải mã. Nhưng tỷ lệ nén của RLE dựa trên cùng một luồng dữ liệu đầu vào liên tiếp và nó không phù hợp với dữ liệu không có nhiều lần chạy, vì nó có thể làm tăng kích thước dữ liệu. Theo cách này, nó cần chuyển đổi dữ liệu trước RLE để có được kết quả nén tốt hơn. [4]







**_Phương pháp nén_**



* Đọc ảnh đầu vào
* Mã hóa ảnh theo từng dòng cho đến khi không còn dòng nào nữa. Đối với mỗi dòng:
    * Với từng bộ giá trị lặp lại, thêm một mục đầu vào trong ma trận được mã hóa ở vị trí thứ i
    * Đếm số lần lặp, lưu giá trị này ở vị trí i+1
    * Dừng lại khi hết dòng
* Áp dụng cho tất cả các dòng

```python
#hàm mã hóa RLE 
#ar là hàng dòng được nén
#chạy từ đầu hàng đến cuối hànd
#Để có kết quả tốt hơn em coi độ khác biệt tuyệt đối giữa các pixel <=10 là giông nhau
#lưu kích thước và dữ liệu mã hóa dưới dạng nhị phân
def run_length_encode( ar ):
    i =0
    s = ""
    while i<len(ar):
        c=0
        el = ar[i]
        while i<len(ar) and  abs(ar[i]-el)<=10 :
            c+=1
            i+=1
        
        el_bin = "{0:08b}".format(el)
        c_bin = "{0:08b}".format(c)
        s = s + el_bin + c_bin
    return s
```

**_Tỉ lệ nén_**

Tỷ lệ nén được sử dụng dữ liệu để ước tính hiệu suất của thuật toán nén. Hiệu suất của thuật toán nén thường được tính toán bằng cách sử dụng tỷ lệ nén được xác định là: 

$$ ratio= \frac{ initial size- compressed size}{ initial size} $$


Trong đó compressedsize và initialsize là kích thước của luồng bit được nén và ban đầu.

```python
#hàm mã hóa toàn bộ ảnh
#Đọc ảnh đầu vào
#Nếu là ảnh màu thì chuyển về ảnh xám
#Mã hóa theo từng dòng theo RLE
#filename: tên tệp txt sẽ được sử dụng để ghi các bit đã chuyển đổi
#=> để chuyển đổi từng hàng thành từng bit và ghi nó vào tệp txt
#Tính % nén
def compress_image(img_ar, filename):
    file = open(filename,"w")
    W, H = img_ar.shape
    bits=0
    for row in img_ar:
        l = run_length_encode(row) + '\n'
        bits+=len(l)
        file.write(l)
    file.close()
    orig_bits = W*H*8
    print("========== COMPRESSION SUCCESSFULLY DONE =========")
    print("COMPRESSED IMAGEFILE NAME : " , filename)
    print("ORIGINAL SIZE OF IMAGE IN BITS (WxHx8)", orig_bits , ' WHICH IS : ' , orig_bits/8000 ,' KB')
    print("AFTER RLE COMPRESSION SIZE IN BITS " ,  bits , ' WHICH IS : ' , bits/8000 , ' KB')
    print("COMPRESSION PERCENTAGE : " , ((orig_bits-bits)/orig_bits)*100 , "%")
```

**_Phương pháp giải nén_**

* Đọc nhị phân từ tệp txt
* Chuyển đổi thành số thập phân và truy xuất phần tử có số đếm
* Truy xuất mảng đến cuối ảnh
* Hiển thị hình ảnh đã giải mã 

```python
def decode( filename ):
    file = open(filename,"r+") 
    extract = file.read().split('\n')
    ar= []
    for li in extract:
        for i in range(0,len(li),16):
            el = li[i:i+8]
            cnt = li[i+8:i+16]
            el = int(el,2)
            cnt = int(cnt,2)
            ar = ar+ cnt*[el]
    
    file.close()
    new_img = np.array(ar)
    new_img = np.reshape(new_img , (W, H))
    new_img = new_img.astype('uint8')
    new_img = Image.fromarray(new_img, mode='P')
    return  new_img
```

Có thể thấy ở trên khi quay ngang ảnh, tỉ lệ nén là cao hơn. Điều này là do trong RLE, số hàng trong hình ảnh được mã hóa giữ nguyên bằng số hàng trong hình ảnh gốc, trong khi số lượng cột thay đổi.
